## Колебание струны

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class Lab6(object):
    def __init__(self, length, number_of_shiftes, a=0.01, k=0.2, h=0.05,shift=None):
        self._length = length
        self._number_of_shiftes = number_of_shiftes
        self.a = a**2
        self.k = k**2
        self.h = h**2
        self.values = np.zeros((self._number_of_shiftes, self._length))
        if shift is None:
            for i in range(self._length):
                self.values[0, i] = float(input())
        else:
            self.values[0, :] = shift
            self.values[1, :] = shift
    
    def __calculate_value(self, i, j):
        first_step = (self.a*self.k*(self.values[i,j+1]-2*self.values[i,j] + self.values[i, j-1]))/self.h
        second_step = 2.*self.values[i, j] - self.values[i-1, j]
        self.values[i+1, j] = first_step+second_step

    def solve(self):
        for i in range(1, self._number_of_shiftes-1):
            for j in range(1, self._length-1):
                self.__calculate_value(i, j)
                
    @staticmethod
    def print_matrix(matrix, message):
        print(message)
        data = pd.DataFrame(matrix)
        display(data)
    
    def plot_results(self):
        plt.figure(dpi=300)
        for i in range(self._number_of_shiftes):
            plt.plot(self.values[i, :])
        plt.show()
    
    def main(self):
        self.solve()
        self.plot_results()
        self.print_matrix(self.values, "Результат")
```

```python
T = 40.
P = 10000.
S = 0.0025
a = (T / (P * S)) ** 0.5
h = 0.1
k = 0.01
length = 5
number_of_shiftes = 52
shift = np.array([0., 0.01, 0.015, 0.01, 0.])
test = Lab6(length, number_of_shiftes, a, k ,h ,shift)
test.main()
```
